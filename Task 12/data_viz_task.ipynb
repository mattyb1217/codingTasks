```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load data
df= pd.read_csv('Cars93.csv')
```

### 1. Boxplot

```python
# Boxplot for revs per mile
plt.figure()
sns.boxplot(x= 'Manufacturer', y= 'Rev.per.mile', data= df)
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


 The car with the highest revs per mile is a Geo

### 2. Histogram

```python
# Histogram of MPG
plt.figure()
sns.histplot(x= 'MPG.city', data= df)
plt.show()

plt.figure()
sns.histplot(x= 'MPG.highway', data= df)
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:** It is more fuel efficient to drive a car on the highway compared to in the city.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure()
sns.lineplot(x= 'Wheelbase', y= 'Turn.circle', data= df)
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
The graph shows a positive correlation between Wheelbase size and the urn circle of the vehicles. A larger wheelbase means a larger turn circle too.

### 4. Barplot

```python
# Barplot for mean horsepower
plt.figure()
sns.barplot(x= 'Horsepower',y= 'Type', data= df)
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:** A larger car doesn't necessarily mean more horsepower but the highest horsepower does come from a lrager vehicle. Midsize and sporty also have quite high horsepowers.
